```python
def baca_konfigurasi(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()

    dfa = {}
    transitions = {}
    for line in lines:
        line = line.strip()
        if line.startswith("States:"):
            dfa['states'] = [s.strip() for s in line[len("States:"):].split(",")]
        elif line.startswith("Alphabet:"):
            dfa['alphabet'] = [s.strip() for s in line[len("Alphabet:"):].split(",")]
        elif line.startswith("Start:"):
            dfa['start'] = line[len("Start:"):].strip()
        elif line.startswith("Accept:"):
            dfa['accept'] = [s.strip() for s in line[len("Accept:"):].split(",")]
        elif line.startswith("Transitions:"):
            continue
        elif line:
            parts = line.split()
            if len(parts) == 3:
                state_from, symbol, state_to = parts
                transitions[(state_from, symbol)] = state_to
    dfa['transitions'] = transitions
    return dfa

harga = {
    'A': 3000,
    'B': 4000,
    'C': 6000
}

def main():
    dfa = baca_konfigurasi("/content/badot-vm.txt")

    current_state = dfa['start']
    lintasan = [current_state]
    total_uang = 0
    max_uang = 10000

    print("Simulasi Vending Machine")
    while True:
        inp = input("Masukkan uang atau beli minuman (1000, 2000, 5000, 10000, A, B, C): ").strip().upper()

        # Jika input adalah pilihan minuman
        if inp in ['A', 'B', 'C']:
            if total_uang == harga[inp]:
                print(f"Lintasan DFA: {' → '.join(lintasan)}")
                print(f"Minuman {inp} dapat dibeli. Status: ACCEPTED.")
                break
            elif total_uang < harga[inp]:
                print(f"Lintasan DFA: {' → '.join(lintasan)}")
                print("Uang tidak cukup. Status: REJECTED.")
                break
            else:
                # Bonus: fitur kembalian
                selisih = total_uang - harga[inp]
                print(f"Lintasan DFA: {' → '.join(lintasan)}")
                print(f"Minuman {inp} dapat dibeli. Status: ACCEPTED.")
                print(f"Kembalian: {selisih}")
                break
        # Jika input adalah uang
        elif inp in dfa['alphabet']:
            nominal = int(inp)
            if total_uang + nominal > max_uang:
                print("Jumlah uang melebihi batas maksimum!")
                continue
            key = (current_state, inp)
            if key in dfa['transitions']:
                current_state = dfa['transitions'][key]
                lintasan.append(current_state)
                total_uang += nominal
                print(f"Total uang: {total_uang}")
                possible = []
                for minuman, harga_min in harga.items():
                    if total_uang >= harga_min:
                        possible.append(f"Minuman {minuman}")
                if possible:
                    print("ON:", ", ".join(possible))
            else:
                print("Transisi tidak valid. Uang yang dimasukkan tidak dapat diterima.")
        else:
            print("Input tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```
